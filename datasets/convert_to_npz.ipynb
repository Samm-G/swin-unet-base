## Convert npy.h5 to npz slice format (For test set..)

```python
import h5py
import os 
import numpy as np

data_dir = 'E:\\Projects\\Data-Science\\MTech\\Courses\\16.Capstone-Projects\\Swin-Unet\\data\\project_TransUNet\\data\\Synapse\\test_vol_h5'
save_dir = 'E:\\Projects\\Data-Science\\MTech\\Courses\\16.Capstone-Projects\\Swin-Unet\\data\\project_TransUNet\\data\\Synapse\\test_npz'
data_dir_list = open('E:\\Projects\\Data-Science\\MTech\Courses\\16.Capstone-Projects\\Swin-Unet\\lists\\lists_Synapse\\test_vol_h5.txt').readlines()
sample_list = open('E:\\Projects\\Data-Science\\MTech\Courses\\16.Capstone-Projects\\Swin-Unet\\lists\\lists_Synapse\\test.txt').readlines()

for idx in range(len(data_dir_list)):
    vol_name = data_dir_list[idx].strip('\n')
    filepath = data_dir + f"\\{vol_name}.npy.h5"
    data = h5py.File(filepath)
    image, label = data['image'][:], data['label'][:]

    if image.shape == label.shape: # Just a safety check..
        for i in range(image.shape[0]):
            data_save_path = os.path.join(save_dir,f'{vol_name}_slice{str(i).zfill(3)}.npz')
            print(f'Saving to: {data_save_path}')
            final_save_slice = {'image':image[i],'label':label[i]}
            with open(data_save_path, 'wb') as f:
                np.savez(f, image=image[i], label=label[i])

    # data_path = os.path.join(data_dir, slice_name+'.npz')
    # print(image)
    # print(label)
    # print(filepath)

# sample_list
```
